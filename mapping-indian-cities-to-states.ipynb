```python
cities_db_loc = f'/kaggle/input/indian-cities-database/Indian Cities Database.csv'
cities_db = pd.read_csv(cities_db_loc)
```

```python
cities_db
```

```python
cities_db.drop('iso2', axis=1)

if 'iso2' in cities_db.columns:
    cities_db.drop('iso2', axis=1, inplace=True)
```

```python
cities_db
```

```python
cities_db.drop('country', axis=1)

if 'country' in cities_db.columns:
    cities_db.drop('country', axis=1, inplace=True)
```

```python
cities_db
```

```python
cd_df = pd.read_csv('/kaggle/input/analyzing-credit-card-spending-habits-in-india/Credit card transactions - India - Simple.csv')
```

```python
cd_df
```

```python
# split the city column to city and country columns

cd_df[['City', 'Country']] = cd_df['City'].str.split(',', expand=True)
cd_df
```

```python
merged_df = pd.merge(cd_df, cities_db, how='left', on='City')
```

```python
# For example, fill missing state values with 'Unknown'
merged_df['State'].fillna('Unknown', inplace=True)
```

```python
merged_df
```

```python
merged_df.drop('index', axis=1)

if 'index' in merged_df.columns:
    merged_df.drop('index', axis=1, inplace=True)
```

```python
#cleaned data set export
df1=merged_df
df1.to_csv('cleaned_data_india_credit_2.csv')
```
